The physiology of sleep dictates that sleep isn't just a state where wakefulness and consciousness are absent, but a more complex and active state.

First off, sleep is not an absent state by any means, in order to initiate sleep there is activity that must occur in order to suppress activity in the ascending arousal system. The inhibatory cells that allow this to happen come from the centrolateral preoptic area (VLPO). VLPOremains active all throught sleep. in order for these inhibitors to take effect and suppress activity there must be a trigger to activate  VLPO, but based on research done by David W. Carley and Sarah S. Farabi in their article ( Physiology of sleep, 2016), "The molecular "triggers" that activate VLPO and initiate sleep onset have not been fully defined but a substantial body of evidence points to extracellular adenosine as a candidate." This research does not pont towards extracellular adenosine as the trigger for sleep and maintaining yourslef in a sleep state which means that there are other factors that could initiate sleep. Another potential culprit to sleep initiation is the monoaminergic arousal centers that project to the VLPO. These two factors alone have complex roles in initiating sleep and mainting the state of being asleep. This shows how sleep is not just an absent state of wakefulness, but shows thath it is a process that must be triggered and activley maintained.

Sleep obviously has a process in which you must initiate sleep but it also actviley goes through a process that helps maintian the state of being asleep. As stated before , sleep is not an absent state of wakefluness, even while asleep there are two fundamentally distinct types of sleep which are: Rapid eye movement (REM) sleep and Non Rapid eye Movement (NREM) sleep. REm sleep is a state of sleep which is associated with dreaming and even though the same type of activity doesnt happen during NREM there is is activity as the the body switches from REM sleep to NREM sleep. According to David W. Carley and Sarah S. Farabi in the article (Physiology of sleep, 2016), "switches between NREM and REM sleep appear to be controlled by reciprocal inhibition between monoaminergic neurons and a specific subset of cholinergic neurons within the brainstem," Also stated in the article, "when REM sleep is triggered, REM-on cholinergic neurons become maximally active, while noradrenergic and serotonergic neurons become virtually silent. The switching between activity and inhibition of these neurons results in a characteristic cycling between NREM and REM during the sleep period." even tho Noradrenergic and serotonergic neurons are almost completely silent, REM-on Cholinergic neurons become maximally active showing that there isw still activity present.

Even with factual evidence of activity occuring in the brain while asleep, people still believe that sleep is a state in which all activity and wakefulness is absent. based on the article written by David W. Carley and Sarah S. Farabi in the article (Physiology of Sleep, 2016) they state, " Typically, sl=keletal muscle activity exhibits progressively decreasing amplitude with transitions from wakefulness to N1, N2, and N3 sleep." Which does state that the level of activity throughout sleep will fall to a state of almost no activity at all. However, as previously stated, throughout sleep there is a constant switching from REM sleep to NREM sleep. With REM sleep the brain and even muscles display activity, while in the case of NREM sleep only the brain shows signs of activity. The activity may be very low during NREM sleep but that does not discount the little brain activity that is present during this state of sleep . Another arguement that has been made to prove that sleep is an absent state of wakefulness is that certain parts of the brain are shutdown during REM sleep. According to David Khan and Trivia Grover in the article (Consciousness in Dreams, 2010), "At least teo importnat regions, the dorsal lateral prefrontal cortex (DLPFC) and the precuneus in the parietal lobe, are deactivated during rapid eye movement (REM) sleep, the period when most dreaming takes place." Also the article states," Asleep and dreaming, the senses are muted if not shut down." Although these examples may give some merit to the idea that sleep is not an active state for the body, studies and evidence shows that even the deepest states of sleep where senses are completely absent the mind is still working and active.

The final peice of information that will provide evidence that sleep is not just an absent state of wakefulness but a state in which the body is still active is the laboratory Polsomnogram(PSG). A PSG is the Physiological gold standard of assessing sleep and awake states.A PSG is conducted by attaching a series of noninvasive sensors to a subject. Two od the readings that are studies through PSG are electroencephalogram (EEG) and electrocardiography (ECG) along with other sensors that monitor abdominal strain, oral and nasalairflow,as well as a finger probe to meaure areterial oxygen saturation. Current PSG's are scored by three stages, stages are N1 - N3, of deepening NREM sleep along with wakefulnessand REM sleep. These readings are seen as waves. The waves are: awake state or alpha waves, stage N1 or theta waves, stage N2 or spindles and K-complex, finally stage N3 or delta waves. The image shows what the waves look like.
